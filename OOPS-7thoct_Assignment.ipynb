Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

```python
class bank_account:
  '''Enter account number,Name'''
  def __init__(self,acc_no,acc_hldr_name):
    self.acc_no=acc_no
    self.acc_hldr_name=acc_hldr_name
    self.acc_bal=0

  def deposit_amt(self,deposit_amt):
    self.acc_bal=self.acc_bal+ deposit_amt
    return self.acc_bal
  def Withdraw_amt(self,wd_amt):
    if wd_amt<self.acc_bal:
      self.acc_bal= self.acc_bal- wd_amt
    else:
      print("Insufficient Fund")
    return self.acc_bal

```

Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

```python
class Shape:
    def __init__(self, length, width=None, height=None):
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        if self.width is None and self.height is None:
            return self.length ** 2
        elif self.height is None:

            return self.length * self.width
        else:

            return 2 * (self.length * self.width + self.length * self.height + self.width * self.height)  # Surface area of a cube

    def calculate_perimeter(self):
        if self.width is None and self.height is None:

            return 4 * self.length
        elif self.height is None:

            return 2 * (self.length + self.width)
        else:

            return 4 * (self.length + self.width + self.height)


square = Shape(5)
rectangle = Shape(4, 6)
cube = Shape(3, 3, 3)

print("Square:")
print("Area:", square.calculate_area())
print("Perimeter:", square.calculate_perimeter())

print("\nRectangle:")
print("Area:", rectangle.calculate_area())
print("Perimeter:", rectangle.calculate_perimeter())

print("\nCube:")
print("Surface Area:", cube.calculate_area())  # For cube, we calculate surface area instead of area
print("Perimeter:", cube.calculate_perimeter())  # Perimeter for a cube is the same as its surface area
```

Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
class todolist:
  def __init__(self):
    self.tasks={}
  def add_tasks(self,newtask,duedate):
    self.tasks[newtask]=duedate
  def mark_task(self,task):
    self.tasks.pop(task)
  def pending_tasks(self):
    return self.tasks
```
