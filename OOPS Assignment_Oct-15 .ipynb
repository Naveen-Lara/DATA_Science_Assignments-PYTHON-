**10. How can you have multiple constructors in a Python class? Explain with an example.**

*   It is not possible to have multiple constructors using different signatures.
*   It is possible by using *Class method or factory methods*,can be used like constructor itself,parameters are passed during instatiation.



````python
class params:
  def __init__(self,radius):
    self.radius=radius
    self.pi=3.141
  @classmethod # constructor of different sign using class method
  def from_diameter(cls,diameter):
    radius= diameter/2
    return cls(radius)
  def calc_area(self):
    area= self.pi*(self.radius**2)
    return area
````

````python
#@title 5. Create a Python class called `Person` with a private attribute `__name`. Provide methods to get and set the name attribute.
class person:
  def __init__(self,name):
    self.__name=name
  def get(self):
    return self.__name
  def set(self,new_name):
    self.__name= new_name
    return self.__name

sudh=person('Sudhanshu kumar')
sudh.get(), sudh.set("Krish Naik")
````

````python
#@title 6. Explain the purpose of getter and setter methods in encapsulation. Provide examples
#@markdown Getter and Setter methods are mainly used in accessing the private datamembers of class,It can be acheived through normal functions as well as through Decorators, using *Property dectorator*

class person:
  def __init__(self,name):
    self._name=name
  @property
  def name(self):
    return self._name
  @name.setter
  def name(self,new_name):
    self._name= new_name
    return self._name

sudh=person('Sudhanshu kumar')
sudh.name # prints sudhanshu kumar, used getter method
sudh.name="Krish Naik"
sudh.name # prints Krish Naik, used setter method
````

````python
#@title 20. Create a Python class for a social media application, using composition to represent users, posts, and comments
class User:
    def __init__(self, username):
        self.username = username
        self.posts = []
        self.comments = []

    def create_post(self, content):
        post = Post(content, self)
        self.posts.append(post)
        return post

    def add_comment(self, post, content):
        comment = Comment(content, self, post)
        self.comments.append(comment)
        post.add_comment(comment)
        return comment

class Post:
    def __init__(self, content, author):
        self.content = content
        self.author = author
        self.comments = []

    def add_comment(self, comment):
        self.comments.append(comment)

class Comment:
    def __init__(self, content, author, post):
        self.content = content
        self.author = author
        self.post = post


user1 = User("Alice")
user2 = User("Bob")


post1 = user1.create_post("Hello, this is my first post!")


comment1 = user2.add_comment(post1, "Nice post, Alice!")


print(f"User: {user1.username}")
print(f"  Post: {post1.content}")
print(f"    Comments:")
for comment in post1.comments:
    print(f"      - {comment.author.username}: {comment.content}")

print(f"\nUser: {user2.username}")
print(f"  Comment: {comment1.content}")
print(f"    Post: {comment1.post.content}")
print(f"    Author of post: {comment1.post.author.username}")
````
